Rock Paper Sissors

```python
import random
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''
game=[rock,paper,scissors]
choose=int(input("what do you chose? Type 0 for Rock,1 for Paper, 2 For Scissors: "))

if choose>=3 or choose<0:
  print("You choosed an Invalid Number")
else:
  print(game[choose])
  computer_chose=random.randint(0,2)
  print("Computer Chooses:")
  print(game[computer_chose])

  if choose==0 and computer_chose==2:
    print("You win!!")
  elif computer_chose==0 and choose==2:
    print("you lose!!")
  elif(computer_chose>choose):
    print("you lose!!!")
  elif(choose>computer_chose):
    print("you win!!!")
  elif(choose==computer_chose):
    print("Draw!!!")
    
print("Thank you Play again!!")
```
